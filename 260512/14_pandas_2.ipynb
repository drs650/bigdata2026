## 불(bool) 인덱싱

```python
w = df['temp'] >= 30
df[w] # 참(True)인것만 행 나온다.
```

### 조건에 맞는 데이터 추출

```python
# 최고로 더웠던 날의 모든 정보(행)
w = df['temp'] == df['temp'].max()
df[w]
```

## 조건이 2개 이상 --> 그리고 &, 또는 |

```python
# 기온이 30도 이상이고, 최대풍속이 9이상인 데이터

w = (df['temp'] >= 30) & (df['max_wind'] >= 9)
df[w]
```

```python
# 기온이 30도 이상이거나, 최대풍속이 9이상인 데이터

w = (df['temp'] >= 30) | (df['max_wind'] >= 9)
df[w]
```

## 결측데이터 대체하기 --> fillna()

```python
df['max_wind'].isnull().sum()
```

```python
df3 = df.copy() # 복사
df3
```

```python
df3 = df3.fillna(0) # 결측치를 모두 0으로 채웠다.
```

```python
df3.isnull().sum()
```

```python
# 평균 --> mean()
avg = df3['max_wind'].mean()
avg
```

```python
df3['max_wind'] = df3['max_wind'].fillna(avg)
```

```python
avg2 = df3['mean_wind'].mean()
avg2
```

```python
df3['mean_wind'] = df3['mean_wind'].fillna(avg2)
```

```python
df3.tail(10)
```

## 컬럼(열, 변수, 피처) 삭제 / 생성

### drop(컬럼 삭제)

```python
dust = pd.read_excel('dust1.xlsx')
dust.head()
```

```python
# 지역, 망, 측정소코드 삭제

dust = dust.drop(['지역', '망', '측정소코드'], axis=1)
```

```python
dust.head()
```

## 컬럼 생성하기(추가)

- 데이터프레임명[추가하고 싶은 컬럼명] = 값

```python
dust['city'] = '서울'
```

```python
dust.head()
```

## 컬럼 이름 수정 --> rename()

- 데이터프레임명.rename(columns={'기존이름' : '새이름'})

```python
dust = dust.rename(columns={
    '측정소명' : 'name', 
    '측정일시' : 'date',
    '주소' : 'addr'
})
```

```python
dust.columns
```

## 데이터 형 확인 --> dtypes

```python
dust.dtypes
```

## 데이터 형 변환 --> astype()

- 컬럼명.astype(바꿀 자료형)

```python
dust['date'].astype(str)
```

```python
dust['date'] = dust['date'].astype(str)
```

```python
dust.dtypes
```

```python
# 첫 8자리만 자른다. --> slice()
dust['date'] = dust['date'].str.slice(0,8)
```

## to_datetime() --> 날짜 / 시간 형으로 변환 (시계열)

```python
pd.to_datetime(dust['date'])
```

```python
dust['date'] = pd.to_datetime(dust['date'])
dust.dtypes
```
